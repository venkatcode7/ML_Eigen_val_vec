```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
```

```python
def plot_vector(ax, vector, color, label):
    arrow = FancyArrowPatch((0, 0), vector, color=color, mutation_scale=15, label=label)
    ax.add_patch(arrow)
    ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color=color)
    ax.text(vector[0], vector[1], label, fontsize=12, color=color, ha='right', va='bottom')

def apply_transformation(matrix, vector):
    return np.dot(matrix, vector)

def plot_transformation(ax, matrix, vector, color_before, color_after):
    plot_vector(ax, vector, color_before, "Before")
    transformed_vector = apply_transformation(matrix, vector)
    plot_vector(ax, transformed_vector, color_after, "After")
    ax.set_aspect('equal', adjustable='datalim')
```

```python
# Example transformation matrices
transformation_matrices = [
    np.array([[2, 1],
              [0, 1]]),
    np.array([[1, -1],
              [1, 1]]),
    np.array([[0, -1],
              [1, 0]])
]
```

```python
# Example vector
input_vector = np.array([1, 2])
```

```python
# Create a plot for each transformation graphical representation
for i, matrix in enumerate(transformation_matrices):
    fig, ax = plt.subplots()
    ax.set_title(f'Transformation Matrix {i+1}')
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    plot_transformation(ax, matrix, input_vector, 'blue', 'red')

plt.show()
```
